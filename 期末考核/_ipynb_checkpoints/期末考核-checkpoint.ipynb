#### 编程基础1：有四个数字：3、5、7、9，能组成多少个互不相同且无重复数字的三位数？输出所有的排列？（10分）

```python
#编程基础1答案写在此单元格内
numbers = [3, 5, 7, 9]              # 初始列表
result = []                         # 存储结果的列表
# 三个for循坏选择三个数字
for i in numbers:
    for j in numbers:
        for k in numbers:
            a = i*100 + j*10 + k    # 组成三位数
            # 判断是否有重复的数值在列表里,没有就添加进去
            if a not in result:
                result.append(a)    # 将未重复的数据添加到result列表中
print(f"结果为{result}，共有{len(result)}个不同的三位数")

            
```

#### 编程基础2：随机生成一个包含20个取值在[1,10]之间的随机整数的列表，将其中的偶数按升序排列放置在列表前部，奇数按降序排列放置在列表后部（10分）

```python
#编程基础2答案写在此单元格内
import numpy

sum_list = np.random.randint(1, 11, 20)   # 生成20个1-10的一维数组

even_list = []  # 偶数列表
old_list = []   # 奇数列表

for i in sum_list:
    if i % 2 == 0:   # 判断是否是偶数
        even_list.append(i)             # 加入偶数列表
    else:            # 判断是否是奇数
        old_list.append(i)              # 加入奇数列表
        
a = np.sort(even_list)         # 偶数列表升序
b = np.sort(old_list)[::-1]    # 奇数列表降序（现正向排序在逆序）
result = []         # 存储结果代码

# 使用两个for循环将数据合并在一起
for i in range(len(a)):
    result.append(a[i])
for i in range(len(b)):
    result.append(b[i])
    
print(f"输出排序结果{result}")      # 输出结果
```

```python
# 代码和分析步骤及结果解释写在之后的单元格中，可以使用多个单元格
# 1. 用户消费金额、消费次数的描述性统计
import numpy as np
import pandas as pd        # 导入模块
import matplotlib as mpl
from matplotlib import pyplot as plt

# 实现图表中文字符显示的必有代码语句
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

# 读取文件
df = pd.read_csv('kzpart.csv')

data = df.groupby(df['user_id'])            # 按照用户分组
x = np.sort(data.count()['price'])         # 统计每个user_id的消费次数
y = np.sort(data.sum()['price'])            # 统计每个user_id的消费金额

plt.plot(x,y)                      # 绘制折线图
plt.title('用户消费金额、消费次数的联系')   # 标题
plt.xlabel('消费次数')             # 横坐标标签
plt.ylabel('消费金额')             # 纵坐标标签
plt.show()
```

```python
# 2.用户消费次数、消费金额散点图

import numpy as np
import pandas as pd      # 导入模块
import matplotlib as mpl
from matplotlib import pyplot as plt

# 使用中文必须加上的代码
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

# 读取文件
df = pd.read_csv('kzpart.csv')
# 按照用户user_id分组
data = df.groupby(df['user_id'])

x = data.count()['price']   # 统计消费次数
y = data.sum()['price']     # 统计消费金额

plt.scatter(x,y, s=10)      # 散点图
plt.title('用户消费次数、消费金额散点图')  # 标题
plt.xlabel('消费次数')             # 横坐标标签
plt.ylabel('消费金额')             # 纵坐标标签
plt.show()
```

```python
# 3.用户消费金额分布图(直方图)
import numpy as np
import pandas as pd      # 导入模块
import matplotlib as mpl
from matplotlib import pyplot as plt

# 使用中文必须加上的代码
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

# 读取文件
df = pd.read_csv('kzpart.csv')
# 按照用户user_id分组
data = df.groupby(df['user_id'])
a = data.agg({'price':'sum'})         # 统计分组后每个用户的消费总额 
print(f"个人最高消费：{a.max()}，个人最低消费：{a.min()}")

x = a.values    # 消费数据列表
bin = np.arange(0,10000,1000)  # 直方图个数和区间

plt.hist(x,bins=bin,histtype='bar')   # 绘制直方图

plt.title('用户消费金额分布图')    # 标题
plt.xlabel('消费金额')             # 横坐标标签
plt.ylabel('消费人数')             # 纵坐标标签
plt.xticks(range(0,10000,1000))    # 横坐标刻度
plt.show()
```

```python
# 4.用户消费次数分布图(直方图)
import numpy as np
import pandas as pd      # 导入模块
import matplotlib as mpl
from matplotlib import pyplot as plt

# 使用中文必须加上的代码
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

# 读取文件
df = pd.read_csv('kzpart.csv')

# data 数据存储用户id和购买次数
data = df['user_id'].value_counts()

print(f'用户购买最多次数为{max(data.values)}，最少次数为{min(data.values)}，中位数为{np.median(data.values)}')
# 条形柱体数量
bins = np.arange(0,20,1)

# x 购买不同次数的人数分布列表
x = data.values

# 绘制直方图，histtype=‘bar’是代表使用条形图
plt.hist(x,bins=bins,  
        color='green',  
        histtype='bar')

plt.title('用户消费次数分布图')     # 标题
plt.xlabel('购买次数')       # 横坐标标签
plt.ylabel('购买人数')       # 纵坐标标签

plt.xticks(range(1,20,5))    # 横坐标刻度
plt.show()
```

```python
# 5.用户累计消费金额占比
import numpy as np
import pandas as pd      # 导入模块
import matplotlib as mpl
from matplotlib import pyplot as plt

# 使用中文必须加上的代码
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

# 读取文件
df = pd.read_csv('kzpart.csv')
# 按照用户user_id分组
data = df.groupby(df['user_id'])

y = data.sum()['price']     # 统计每个user_id的消费金额

y.count()                                     # 统计总的购买人数
y1 = (y.values < 500.00).sum()                # 消费额<500的人数
y2 = (y.values < 1000.00).sum() - y1          # 500<消费额<1000的人数
y3 = (y.values < 1500.00).sum() - (y2+y1)     # 1000<消费额<1500的人数
y4 = (y.values < 5000.00).sum() - (y3+y2+y1)  # 1000<消费额<5000的人数
y5 = (y.values > 5000.00).sum()               # 消费额>5000的人数


elements = ["消费额<500","500<消费额<1000","1000<消费额<1500","1000<消费额<5000","消费额>5000"]  # 标签列表
weights = [y1/y.count(),y2/y.count(),y3/y.count(),y4/y.count(),y5/y.count()]                     # 数据百分比列表
colors = ["#8dd3c7","#377eb8","#4daf4a","#984ea3","#ff7f00"]                                     # 颜色列表

plt.pie(weights,              # 输入的百分比数据
       labels = elements,     # 标签列表
       autopct = "%3.1f%%",   # 控制百分比的位数
       colors=colors)         # 颜色列表

# 通过legend函数中的loc和bbox_to_anchor参数控制图例位置
plt.legend(elements,title = "用户累计消费金额占比",bbox_to_anchor = (1,0,0.3,1))


plt.show()
```
