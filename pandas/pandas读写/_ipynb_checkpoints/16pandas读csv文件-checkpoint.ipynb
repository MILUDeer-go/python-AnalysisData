```python
# to_string() 用于返回 DataFrame 类型的数据，如果不使用该函数，则输出结果为数据的前面 5 行和末尾 5 行，
# 中间部分以 ... 代替
# read_csv（）函数用来读取csv文件
import pandas as pd 

df = pd.read_csv("nba.csv",nrows=5)
print(df)
print("="*30)
print(df.to_string())
```

```python
# read_csv（）函数用来读取csv文件,文件以逗号为分隔符，同时将其转换为DataFrame对象
# 含有标题的csv文件默认第一行就是列名。
import pandas as pd 

fp1 = pd.read_csv("myCSV_01.csv")
fp1
```

```python
# csv文件被视为文本文件，可以使用read_table（）函数，但是需要指定分隔符

fp2 = pd.read_table("myCSV_01.csv",sep=',')  # 未指定分割符时候，数据会被分为一列
fp2
```

```python
# 下面是第一行开始就是列表数据的情况,对于没有表头的这种情况，一般使用header选项，将其值设置为None即可
fp3 = pd.read_csv("myCSV_02.csv")
print(fp3)
print("="*30)
fp3 = pd.read_csv("myCSV_02.csv",header=None)
fp3
```

```python
# 可以使用names选项设置表头，直接把存有各列名称的数组赋给它即可
pd.read_csv("myCSV_02.csv",names=['white','red','blue','black','animal'])
```

```python
# 使用RegExp解析TXT文件
# 如果解析的数据文件不是以逗号或者分号分隔的。可以使用sep选项指定正则表达式，在read_table（）函数内使用
import pandas as pd 
txt = pd.read_table("ch01.txt")
print(txt)
print("="*30)
# 使用正则表达式分隔
txt1 =  pd.read_table("ch01.txt", sep='\s+')  #使用\s+来代表多个空白字符，s*不适用，所以用s+
txt1 
```

```python
# 在read_table()选项中可以使用skiprows()在读取数据时候移除不需要的行
import pandas as pd 

fd = pd.read_table("myCSV_01.csv",sep=',',skiprows=[1,2,3])   # 移除了前三行
fd
```

```python
# 当文件太大时候，只需要读取部分数据时候就需要迭代
# nba.csv文件太大， 我们只读取其中的10-15行
import pandas as pd 

fd1 = pd.read_csv('nba.csv',skiprows=[10],nrows=5)
fd1
```
