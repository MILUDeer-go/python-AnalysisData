###箱线图是由一个箱体和一堆箱须所组成的统计图形。
箱体是由第一四分位数，中位数（第二四分位数）和第三四分位所组成。在箱须的末端之外的数值可以理解为离群值，
因此，箱须是一对组数据范围的大直观描述
应用场景——多组定量数据的分布比较

```python
# 3.9.1 绘制原理----plt.boxplot()
# -*- conding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.sans-serif"]=["FangSong"]
mpl.rcParams["axes.unicode_minus"]=False

testA = np.random.randn(5000)
testB = np.random.randn(5000)

testlist = [testA,testB]
labels = ['随机数生成器ALphaRM',"随机数生成器BetaRM"]
colors = ["#1b9e77","#d95f02"]

whis = 1.6
width = 0.35

bplot = plt.boxplot(testlist,
                   whis = whis,    # 4分位间距的倍数，用来确定箱须包含数据的范围的大小
                   widths = width,  # 设置箱体的宽度
                   sym = "o",         # 离群值的标记样本
                   labels = labels,
                   patch_artist = True)   # 是否给箱体添加颜色

for patch,color in zip(bplot["boxes"],colors):
    patch.set_facecolor(color)

plt.ylabel("随机数值")
plt.title("生成器抗干扰能力的稳定性的比较")

plt.grid(axis = "y",ls=":",lw=1,color="gray",alpha=0.4)
plt.show()
```

```python
# 3.9.2绘制水平方向的箱线图plt.boxplot()
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.sans-serif"]=["FangSong"]
mpl.rcParams["axes.unicode_minus"]=False

x = np.random.randn(1000)
plt.boxplot(x,vert=False)  # vert控制箱线图是水平方向还是竖直方向，前者为False，后者为True

plt.xlabel("随机数值")
plt.yticks([1],["随机数生成器ALphaRM"],rotation = 90)  # 标签旋转90度
plt.title("随机数生成器抗干扰能力的稳定性")

plt.grid(axis="x",ls=":",lw=1, color="gray",alpha=0.4)

plt.show()
```

```python
# 3.9.4 绘制不离群值的水平放置的箱线图plt.boxplot(x,vert=False)
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.sans-serif"]=["FangSong"]
mpl.rcParams["axes.unicode_minus"]=False

x = np.random.randn(1000)

plt.boxplot(x,vert=False,showfliers=False)  # 后面的showfliers参数为False表示不绘制离群值

plt.xlabel("随机数值")
plt.yticks([1],["随机数生成器AlphaRM"],rotation=90)
plt.title("随机数生成器的抗干扰能力的稳定性")

plt.grid(axis="x",ls=":",lw=1,color="gray",alpha=0.4)
plt.show()
```

```python
# 3.10 绘制误差棒图----plt.errorbar()
# 应用场景是——定量数据的误差范围
# 误差棒图就可以很好的实现充当总体参数估计的置信区间的角色。误差棒图的计算方法可以有很多种：
# 单一数值，置信区间，标准差等。误差棒的可视化展示效果也有很多种样式：水平误差棒，垂直误差棒，
# 对称误差棒和非对称误差棒
```

```python
# 3.10.1 绘制原理---plt.errorbar()
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1,0.6,20)
y = np.exp(x)    # e^x函数

error = 0.05 + 0.15*x
lower_error = error
upper_error = 0.3*error
error_limit = [lower_error,upper_error]

plt.errorbar(x,y,yerr=error_limit,     # yerr:单一数值的非对称形式误差范围
            fmt = "o",ecolor = "y",    # fmt：数据点的标记样式和数据点标记的连接线样式
            elinewidth=2,              # ecolor,elinewidth分别表示误差棒的线条颜色和线条粗细
            ms = 5, mfc = "c", mec="r",  # ms,mc,mec分别表示数据点的大小，颜色，标记边缘颜色
            capthick=1, capsize = 2)     # capthick:误差棒边界横杠的厚度
                                         # capsize:误差棒边界横杠大小

plt.xlim(0,0.7)
plt.show()
```

```python
# 带误差棒的柱状图
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.sans-serif"]=["FangSong"]
mpl.rcParams["axes.unicode_minus"]=False

x = np.arange(5)
y = [100,68,79,91,82]
std_err = [7,2,6,10,5]

error_attri = dict(elinewidth=2,ecolor="black",capsize=3)

plt.bar(x,y,color="c",
       width=0.6,align="center",
       yerr=std_err,   # 设置误差棉棒的关键
       error_kw = error_attri,
       tick_label = ["园区1","园区2","园区3","园区4","园区5"])

plt.xlabel("芒果种植区")
plt.ylabel("收割量")

plt.title("不同芒果种植区的单次收割量")

plt.grid(True,axis="y",ls=":",lw=1,color="gray",alpha=0.2)
plt.show()
```

```python
# 带误差棒的条形图
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt 
import numpy as np

mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

x = np.arange(5)
y = [1200,2400,1800,2200,1600]
std_err = [150,100,180,130,80]

bar_width = 0.6
colors = ["#377eb8","#4daf4a","#984ea3","#ff7f00","#1b9e77"]

plt.barh(x,y,bar_width,
        color = colors,
        align = "center",
        xerr = std_err,
        tick_label = ["家庭","小说","心理","科技","儿童"])

plt.xlabel("订购数量")
plt.ylabel("图书种类")

plt.title("大型图书展览会的不同种类的采购情况")
plt.xlim(0,2600)

plt.grid(axis="x",ls=":",lw=2,color="gray",alpha=0.4)
plt.show()
```

```python
# 带误差棒的多数据并列柱状图
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

x = np.arange(5)
y1 = [100,68,79,91,82]
y2 = [120,75,70,78,85]

std_err1 = [7,2,6,10,5]
std_err2 = [5,1,4,6,7]

bar_width = 0.4

err_attri = dict(elinewidth=2,ecolor="black",capsize=3)  # 误差棒末端横截面的宽度
tick_labels = ["园区1","园区2","园区3","园区4","园区5"]

plt.bar(x,y1,bar_width,
        color = "#87CEEB",
        align = "center",
        yerr = std_err1,
        error_kw = err_attri,
        label = "2010")
               


plt.bar(x+bar_width,y2,
        bar_width,
        color = "#CD5C5C",
        align = "center",
        yerr = std_err2,
        error_kw = err_attri,
        label = "2013")

plt.xlabel("芒果种植区")
plt.ylabel("收割量")

plt.xticks(x+bar_width/2,tick_labels)
plt.title("不同年份的芒果种植区的单次收割量")
plt.grid(True,axis="y",ls=":",lw=2,color="gray",alpha=0.2)

plt.legend()
plt.show()
```

```python
# 带误差棒的堆积柱状图
# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

x = np.arange(5)
x = np.arange(5)
y1 = [1200,680,790,910,820]
y2 = [1200,750,700,780,850]

std_err1 = [70,20,60,100,50]
std_err2 = [50,10,40,60,70]

bar_width = 0.4

err_attri = dict(elinewidth=2,ecolor="black",capsize=0)  # capsize误差棒末端横截面的宽度
tick_labels = ["家庭","小说","心理","科技","儿童"]

plt.bar(x,y1,bar_width,
        color = "#87CEEB",
        align = "center",
        yerr = std_err1,
        error_kw = err_attri,
        label = "地区1")
               
plt.bar(x,y2,
        bar_width,
        bottom = y1,
        color = "#CD5C5C",
        align = "center",
        yerr = std_err2,
        error_kw = err_attri,
        label = "地区2")

plt.xlabel("图书种类")
plt.ylabel("订购数量")

plt.title("不同地区的大型图书馆展销会的图书采购情况")

plt.grid(True,axis="y",alpha=0.2)

plt.legend()
plt.show()
```
