```python
# 1.1.1规则多边形的颜色填充
import matplotlib.pyplot as plt
import numpy as np

# 设置坐标点（x,y）
x = [0,0,5,10,15,15,10,5]  
y = [5,10,15,15,10,5,0,0]

# 将围成八边形用蓝色颜色充满
plt.fill(x, y, color = "cornflowerblue")  

# 设置横纵坐标的最大最小值
plt.xlim(-1, 16)
plt.ylim(-1, 16)

# 设置横纵坐标的刻度
plt.xticks(np.arange(0,16,5))
plt.yticks(np.arange(0,16,5))

# 展现绘制效果
plt.show()
```

```python
# 1.1.2不规则多边形的颜色填充
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi,50)   # 2×np.pi就相当于2π啦，0-2π均匀选择50个数
y = np.sin(x)                    # 函数sin()

plt.fill(x,y,color="cornflowerblue", alpha = 0.4)

plt.plot(x,y,color="cornflowerblue", alpha = 8)   # 借助函数plot（）给填充区添加区域轮廓
plt.plot([x[0],x[-1]],[y[0],y[-1]],x,y,color="cornflowerblue", alpha = 8)

plt.xlim(0,2*np.pi)
plt.ylim(-1.1,1.1)

plt.show()
```

```python
# 1.2  交叉曲线的颜色填充
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,500)
y1 = np.sin(2* np.pi* x)
y2 = 1.1* np.sin(3* np.pi* x)

#fig:表示画布。函数的两个整数参数指定子图网格的行数和列数。该函数返回一个图形对象和一个
#包含等于nrows * ncols的轴对象的元组。每个轴对象都可通过索引访问。
#这里创建一个3行一列的子图，横坐标格式一致
fig, ax = plt.subplots(3,1,sharex = "all")

#在第一个子图绘制,填充曲线y2与0之间的区域
ax[0].fill_between(x, 0,y2, alpha = 0.4)
ax[0].set_ylim(-1.2,1.2)

#在第二个子图绘制，填充曲线y2与1.2之间的区域
ax[1].fill_between(x, y2,1.1, alpha = 0.4)
ax[1].set_ylim(-1.2,1.2)

#在第三个子图绘制，填充曲线y2与y1之间的区域
ax[2].fill_between(x, y1,y2, alpha = 0.4)
ax[2].set_ylim(-1.2,1.2)

plt.show()
```

```python
# 1.3.1水平方向的交叉曲线的颜色填充方法
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,500)
y1 = np.sin(2* np.pi* x)
y2 = 1.1* np.sin(3* np.pi* x)

fig = plt.figure()

# 将画布分为2行2列的子图，在1副子图绘制
ax = fig.add_subplot(221)
plt.plot(x,y1)

# 将画布分为2行2列的子图，在2副子图绘制
ax = fig.add_subplot(222)
plt.plot(x,y2)

# 将画布分为2行2列的子图，在3副子图绘制
ax = fig.add_subplot(223)
plt.plot(x,y1<y2)

# 将画布分为2行2列的子图，在4副子图绘制
ax = fig.add_subplot(224)
ax.plot(x,y1,color="k",lw=1,ls="-")
ax.plot(x,y2,color="k",lw=1,ls="-")

# 当y1<y2时候
ax.fill_between(x,y1,y2,where= y2>y1,interpolate=True,
               facecolor="cornflowerblue",alpha =0.7)

# 当y1>y2时候
ax.fill_between(x,y1,y2,where= y1>y2,interpolate=True,
               facecolor="red",alpha =0.7)  # 参数interpolate=True，表示将曲线交叉点附近的公共区域用指点颜色填充

# 设置刻度
ax.set_xlim(0,2)
ax.set_ylim(-1.2,1.2)

ax.grid(ls=":",lw =1, color="gray", alpha=0.5)  # 设置表格线

plt.show()
```

```python
# 1.3.2垂直方向的交叉曲线的颜色填充方法
import matplotlib.pyplot as plt
import numpy as np

y = np.linspace(0,2,500)
x1 = np.sin(2* np.pi* y)
x2 = 1.1* np.sin(3* np.pi* y)

fig = plt.figure()


# 将画布分为1行1列的子图，在1副子图绘制
ax = fig.add_subplot(224)
ax.plot(x1,y,color="k",lw=1,ls="-")
ax.plot(x2,y,color="k",lw=1,ls="-")

#借助fill_betweenx()来实现垂直交叉方向的交叉曲线的颜色填充的目标
# 当x1<x2时候
ax.fill_betweenx(y,x1,x2,where= x2>x1,interpolate=True,
               facecolor="cornflowerblue",alpha =0.7)

# 当x1>x2时候
ax.fill_betweenx(y,x1,x2,where= x1>x2,interpolate=True,
               facecolor="red",alpha =0.7)  # 参数interpolate=True，表示将曲线交叉点附近的公共区域用指点颜色填充

# 设置刻度
ax.set_xlim(-1.2,1.2)
ax.set_ylim(0,2)

ax.grid(ls=":",lw =1, color="gray", alpha=0.5)  # 设置表格线

plt.show()
```
