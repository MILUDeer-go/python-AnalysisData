```python
# 5.1.1刻度定位器（locator）和刻度格式器（formatter）的使用方法
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator,MultipleLocator,FuncFormatter

x = np.linspace(0.5,3.5,100)
y = np.sin(x)

# 设置画布
fig = plt.figure(figsize  = (8,4))   # 构建对象（画布的大小）
ax = fig.add_subplot(111)    # 创建实例，添加第一块画布

# 设置 x,y 主要刻度定位器
ax.xaxis.set_major_locator(MultipleLocator(1.0))   # 每隔一个单位就是主刻度线（下图加粗的刻度线）
ax.yaxis.set_major_locator(MultipleLocator(1.0))

# 设置 x,y 次要刻度定位器 
ax.xaxis.set_minor_locator(AutoMinorLocator(4))   # 每一个单位设置4个等分区
ax.yaxis.set_minor_locator(AutoMinorLocator(4))

# 设置 x 的次要刻度格式器
def minor_tick(x,pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x

ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))  # FuncFormatter 用来控制精度

ax.tick_params("y",which = 'major',     # which：样式 length:长度
              length = 15,width=2.0,
              colors = 'r')   # 主刻度的具体执行语句

ax.tick_params(which="minor",
              length=5,width=1.0,
              labelsize=10,labelcolor='0.25')   # 次刻度的执行

ax.set_xlim(0,4)
ax.set_ylim(0,2)

# ax.plot(x,y,c=(0.25,0.25,1.00),lw =2,zorder=10)
ax.plot(x,y,c=(0.25,0.25,1.00),lw =2,zorder=0)


#ax.grid(linestyle="-",linewidth=0.5,color='r',zorder=0)
ax.grid(linestyle="-",linewidth=0.5,color='r',zorder=10)
# ax.grid(linestyle="--",linewidth=0.5,color='r',zorder=0)

plt.show()

```

```python
# 5.1.2 案例—刻度标签和刻度样式的定制化
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(facecolor=(1.0,1.0,0.9427))   # 构建对象
ax = fig.add_axes([0.1,0.4,0.5,0.5])

for ticklabel in ax.xaxis.get_ticklabels():    # 对x轴的标签来进行修饰
    ticklabel.set_color("slateblue")
    ticklabel.set_fontsize(18)
    ticklabel.set_rotation(30)

for tickline in ax.yaxis.get_ticklines():  # 对y轴的线条进行修饰
    tickline.set_color("lightgreen")
    tickline.set_markersize(20)
    tickline.set_markeredgewidth(2)
    
plt.show()
```

```python
# 5.1.3 货币和时间序列样式的刻度标签
import matplotlib.pyplot as plt
import numpy as np

from calendar import month_name,day_name  # 导入时间模块
from matplotlib.ticker import FormatStrFormatter

fig = plt.figure()

ax = fig.add_axes([0.2,0.2,0.7,0.7])   # 添加坐标轴生成实例

x = np.arange(1,8,1)
y = 2 * x

ax.plot(x,y,ls="-",lw=2,color="orange",marker="o",ms=20,mfc="c",mec="c")

# 设置人民币标签
ax.yaxis.set_major_formatter(FormatStrFormatter(r"$\yen%1.1f$"))
# 设置日期标签
plt.xticks(x,day_name[0:7],rotation =90)  # rotation 设置回转的角度
plt.show()
```

```python
# 5.2 添加指示注解和无指示注解可以使用注解annotate,相应的面向对象的实例方法使Axes.annotate()
# 有指示注解是通过箭头指示的方法对绘图区域的内容进行解释
# 无指示注解是单纯使用文本进行内容解释或者是说明的标注方法
```

```python
# 5.2.1有指示注解annotate()和无指示注解text()的添加方法
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.5,3.5,100)
y = np.sin(x)

fig = plt.figure(figsize=(8,8))   # 创建对象
ax = fig.add_subplot(111)  # 创建实例

ax.plot(x,y,c="lightblue",ls="--",lw=2)

ax.annotate("maximum",xy=(np.pi/2,1.0),xycoords="data",  # “maximum”是注解内容，xy是解释内容的位置，xycoords=“data”表示与折线图使用相同的坐标
           xytext=((np.pi/2)+0.15,0.8),textcoords="data", # 注释内容所在的位置，textcoords：xytext的坐标系统
           weight="bold",color="r",           # weight：注解内容的显示风格 ，color：内容
           arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color="r"))  #arrowprops:指示箭头的属性

ax.text(2.8,0.4,"$y=\sin(x)$",fontsize=20,color="b",
       bbox=dict(facecolor='y',alpha = 0.5))


plt.show()
```
