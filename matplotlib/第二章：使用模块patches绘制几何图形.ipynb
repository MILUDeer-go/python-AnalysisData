```python
# 2.1椭圆的实现方法
# 通过Ellipse类实现
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse

# 将画布分为两部分
fig, ax = plt.subplots(1,3, subplot_kw = {"aspect": "equal"})  # 简单而言，就是长宽刻度增长一致

# 绘制第一部分画布(131)
angles = np.linspace(0,135,4)  # 将0-135均匀分为4部分来获取角度数组
# 逆时针获取4个椭圆实例列表，也就是4个数组，圆心为（2，2），
ellipse = [Ellipse((2,2),4,2,a) for a in angles]

for elle in ellipse:
    ax[0].add_patch(elle)
    elle.set_alpha(0.4)
    elle.set_color("cornflowerblue")

ax[0].axis([-1,5,-1,5])  # 设置子图一的横纵坐标

# 绘制第二块子图(132)
# 这是一个简单的椭圆，宽度为4，长度为8，角度为90
elle = Ellipse(xy=(0,0), width = 4, height = 8, angle=90,facecolor='blue',alpha=0.4)
ax[1].add_patch(elle)

ax[1].axis([-5,5,-3,3]) 

# 绘制第三块子图（133）
# 通过rand()函数可以返回一个或一组服从“0~1”均匀分布的随机样本值。随机样本取值范围是[0,1)，不包括1。

num = np.arange(0,100,1)  # 这个函数是numpy才有的,range函数python自带，不需要导入模块

ellipse = [Ellipse(xy = np.random.rand(2)*10,
            width = np.random.rand(1), 
            height = np.random.rand(1), 
            angle=np.random.rand(1)*360) for i in num]

           
           
for elle in ellipse:
    ax[2].add_patch(elle)
    elle.set_alpha(np.random.rand(1))
    elle.set_color(np.random.rand(3))

# 设置图表轴长度范围
ax[2].axis([-1,11,-1,11])

# tight_layout会自动调整子图参数，使之填充整个图像区域。这是个实验特性，
# 可能在一些情况下不工作。它仅仅检查坐标轴标签、刻度标签以及标题的部分
plt.tight_layout()   
plt.show()
```

```python
# 矩形的实现方法——通过Rectangle类生成矩形
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

# 绘制画布
fig,ax = plt.subplots(subplot_kw={"aspect":"equal"})

x1 = np.arange(1,2.6,0.1)   # 数组0-2.6间距0.1
y1 = x1 + 2

x2 = np.arange(2.5,4.1,0.1)
y2 = -x2 + 7

# 设置背景颜色
rectangle = ax.patch  # 获得矩形实例
rectangle.set_facecolor("lightskyblue")

# 做房子
# (1,0)表示左下角的的点，3，3表示长宽为3，3的矩形
rectangle1 = Rectangle((1,0),3,3,facecolor='w',edgecolor='rosybrown')

# 做大门
rectangle2 = Rectangle((1.5,0),1,1.5,facecolor="w",edgecolor="rosybrown",hatch="|||") 

# 做窗户
rectangle3 = Rectangle((2.9,1.7),0.6,0.6,facecolor="w",edgecolor="rosybrown")

# 房顶线条（roof line）是用数组来标识点与点之间连线
ax.plot([1,2.5,4],[3,4.5,3],color="rosybrown")

# 绘制窗户线条
ax.plot([2.9,3.5],[2,2],color="rosybrown")
ax.plot([3.2,3.2],[1.7,2.3],color="rosybrown")

rectangle_list = [rectangle1,rectangle2,rectangle3]

# 绘制房顶颜色
ax.fill_between(x1,3,y1,color="w",interpolate=True)
ax.fill_between(x2,3,y2,color="w",interpolate=True)

for rect in rectangle_list:
    ax.add_patch(rect)
    
ax.axis([0,5,0,5])

plt.show()
```

## 总结
1.圆的绘制直接使用Circle((x,y),radius=?)方法,其中的参数是必须的，其余参数视情况而定
2.椭圆的绘制则是使用方法Ellipse(xy=(0,0), width = 4, height = 8, angle=90,facecolor='blue',alpha=0.4)，
其中的必备参数是中心和长度和宽度。
3.矩形的绘制使用Rectangle((c ,d ),a,b)；c,d代表矩形的左下顶点；a,b代表长度和宽度
4.绘制直线时候只需要使用plot()传入横坐标的数组和纵坐标数组，两两配对成为点
