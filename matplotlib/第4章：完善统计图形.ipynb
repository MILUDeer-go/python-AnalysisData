```python
# 4.1 添加图例和标题
# 4.1.1 图例和标签的设置方法

# -*- coding:utf-8 -*-
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

x = np.linspace(-2*np.pi,2*np.pi,200)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label=r"$\sin(x)$")  # label种使用r"$$"的模式，将表达式嵌入美元符号其中，以斜体的形式输出
plt.plot(x,y2,label=r"$\cos(x)$")

plt.legend(loc="lower left")

plt.title("正选函数和余弦函数的折线图")

plt.show()
```

```python
# 4.1.2 图例展示样式的调整
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(0,2.1,0.1)
y1 = np.power(x,3)
y2 = np.power(x,2)
y3 = np.power(x,1)

plt.plot(x,y1,label=r"$x^(3)$")
plt.plot(x,y2,label=r"$x^(2)$")
plt.plot(x,y3,label=r"$x^(1)$")

plt.legend(loc = "upper left", bbox_to_anchor=(0.05,0.95),ncol=3,
          title="power function", shadow=True,fancybox=True)

# legend()的关键字参数主要有位置参数loc,线框位置参数bbox_to_anchor,图例标签内容的标签参数title,
# 线框阴影shadow()和线框圆角处理参数fancybox等
plt.show()
```

```python
# 4.1.3 标题的展示样式的调整
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2,2,10000)
y = np.exp(x)

plt.plot(x,y,ls="-",lw=2,color="g")

plt.title("center demo")

# 设置标题既可以放在字体字典中，也可以不使用字典，family是字体类别，size是字体大小，字体颜色是color
# 字体风格style
plt.title("left demo",loc="left",   # 在图的左边设置一个标题
         fontdict={"size":"xx-large",
                  "color" :"r",
                  "family":"Times New Roman"})

plt.title("right demo",loc="right", # 在图的右边设置一个标题
         family = "Comic Sans MS",
         size = 20,
         style = "oblique",
         color = "c")

plt.show()
```

```python
# 4.1.4 带图例的饼图

# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

elements = ["面粉","砂糖","奶油","草莓酱","坚果"]

weights = [40,15,20,10,15]

colors = ["#1b9e77","#d95f02","#7570b3","#66a61e","#e6ab02"]

wedges,texts,autotexts = plt.pie(weights,
                                autopct = "%3.1f%%",      # 圆内部的文本格式（小数点后一位）
                                textprops = dict(color="w"),
                                colors = colors)

plt.legend(wedges,elements,    # 函数legend()的参数wedges和elements分别表示饼片的实例列表和文本标签列表
          fontsize=12,
          title="配料表",
          loc = "center left",
          bbox_to_anchor = (0.91,0,0.3,1))

plt.setp(autotexts,size=15,weight="bold")
plt.setp(texts,size=12)

plt.title("果酱面包的配料比例表")
plt.show()
```

```python
# 4.2 调整刻度范围和刻度标签
# 刻度范围是绘图区域中坐标轴的取值区间，包括x轴和y轴的取值区间。刻度范围是否合适
# 直接决定绘图区域中的图形效果的优劣。
"""
函数subplot(),这个函数专门用来绘制几何图形的相同的网格区域，子区就是将画布分为若干个子图，
这些子图画布就构成了几何形状规则且对称的矩形绘图区域，然后在这些绘图区域上分别绘制图形。
"""
```

```python
# 4.2.1 调整刻度范围和刻度标签的方法
import matplotlib as mpl
import numpy as np

x = np.linspace(-2*np.pi,2*np.pi,200)
y = np.sin(x)

# 绘制子图211
plt.subplot(211)
plt.plot(x,y)

# 绘制子图212
plt.subplot(212)
plt.xlim(-2*np.pi,2*np.pi)
plt.xticks([-2*np.pi,-3*np.pi/2,-1*np.pi,-1*(np.pi)/2,0,np.pi/2,np.pi,3*np.pi/2,2*np.pi],
          [r'$-2\pi$',r'$-3\pi/2$',r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$'])
# 前面的r‘$$’来解析，r表示该字符串是raw strings;$$符号之间的内容会以斜体的形式输出
plt.plot(x,y)
plt.show()
```

```python
# 逆序设置坐标轴的刻度标签

# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

time = np.arange(1,11,0.5)
machinePower = np.power(time,2)+0.7  # power（）函数求time的2次方

plt.plot(time,machinePower,
        linestyle='-',
        linewidth=1,
        color = 'r')

plt.xlim(10,1)  # 将横坐标设为逆序就可以绘制下降图

plt.xlabel("使用年限")
plt.ylabel("机器功率")
plt.title("机器耗损曲线")

plt.grid(axis='y',ls=':',lw=1,color = 'gray',alpha = 0.3)

plt.show()
```

```python
# 向统计图添加表格（这里使用饼图）

# -*- coding:utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt


mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

labels = 'A难度','B难度','C难度','D难度'    # 设置饼图的标签
students = [0.35,0.15,0.3,0.2]
explode = (0.1,0.1,0.1,0.1)

colors = ["#377eb8","#4daf4a","#984ea3","#4daf4a"]

# 绘制饼图
plt.pie(students,explode=explode,     # 每一部分饼图的偏移量
       labels = labels,autopct = '%1.1f%%',
       startangle = 45,            # 饼图旋转角度
       shadow = True,
       colors=colors)
       
plt.title("选择不同难度测试的学生百分比")
# 添加表格数据
colLabels = ['A难度','B难度','C难度','D难度']
rowlabels = ["学生选择试卷人数"]
studentsValues = [[350,150,300,200]]
colColors = ["#377eb8","#4daf4a","#984ea3","#4daf4a"]


# 绘制表格到饼图
plt.table(cellText = studentsValues,   # 表格里面的数值，列表存储
         cellLoc = "center",           # 表格数据的对其方式 
         colLabels = colLabels,        # 表格的每列列名称
         colWidths = [0.2]*4,          # 每列的宽度
         colColours = colColors,        #  每列列名所在的单元格格式
         rowLabels = rowlabels,
         rowLoc = "center",             # 每行的行名称对其方式
         loc = "bottom")                # 表格在画布的位置

plt.show()
```
