```python
# 3.6.3 用函数step()绘制阶梯图
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(1,10,10)
y = np.sin(x)

# where的参数表示从哪里开始画起，pre是从右向左开始，在最左边的那一个结束
# post则是表示接替从左向右边开始绘制，在最右边结束
plt.step(x,y,color="#8dd3c7",where="pre",lw=2)

plt.xlim(0,11)
plt.xticks(np.arange(1,11,1))
plt.ylim(-1.2,1.2)
plt.title("step book")

plt.show()
```

```python
# 3.8 饼图是用来展示定性数据比例分布特征的统计图形，可以秦楚的观察出数据的占比情况
# 主要用于应用在定性数据的可视化场景中，或者是用来进行离散数据的比例展示
```

```python
# 3.8.1 使用函数pie()来绘制饼图
# -*- coding:utf-8 -*-
import matplotlib as mpl
# 设置字体格式
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

import matplotlib.pyplot as plt
import numpy as np

labels = "A难度","B难度","C难度","D难度"
students = [0.35,0.15,0.20,0.30]
explode = (0.1,0.1,0.1,0.1)
colors = ["#377eb8","#4daf4a","#984ea3","#ff7f00"]

plt.pie(students,
       explode=explode,   # 饼片边缘偏离半径百分比
       labels = labels,
       autopct = "%3.1f%%",   # 饼片文本标签内容对应的数值百分比样式
       startangle = 45,        # 从x轴开始，饼片逆时针旋转45
       shadow = True,    # 是否绘制饼片阴影
       colors = colors)

plt.title("选择不同难度测试的学生比例")

plt.show()
```

```python
# 3.8.2 使用函数pie()来绘制饼图——非分裂式饼图
# -*- coding:utf-8 -*-
import matplotlib as mpl
# 设置字体格式
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

import matplotlib.pyplot as plt
import numpy as np

labels = "A难度","B难度","C难度","D难度"
students = [0.35,0.15,0.20,0.30]
explode = (0.1,0.1,0.1,0.1)
colors = ["#377eb8","#4daf4a","#984ea3","#ff7f00"]

plt.pie(students,
       labels = labels,
       autopct = "%3.1f%%",   # 饼片文本标签内容对应的数值百分比样式
       startangle = 45,        # 从x轴开始，饼片逆时针旋转45
       shadow = True,
       pctdistance = 0.5,     # 控制百分比数值的显示位置
       labeldistance = 1.1,    # 控制标签的显示位置，以半径长度比例值作为位置依据
       colors = colors)

plt.title("选择不同难度测试的学生比例")

plt.show()
```

```python
# 案例——绘制内嵌环行饼图
# 饼图不仅可以以用来描述定性数据的比例分布，还可以将多个饼图进行嵌套，从而
# 实现内嵌饼图的可视化效果

# -*- coding:utf-8 -*-
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 设置中文字体可视
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

elements = ["面粉","砂糖","奶油","草莓酱","坚果"]

weight1 = [40,15,20,10,15]
weight2 = [30,25,15,20,10]

colormapList = ["#8dd3c7","#377eb8","#4daf4a","#984ea3","#ff7f00"]
outer_colors = colormapList
inner_colors = colormapList

wedges1, texts1, autotexts1 = plt.pie(weight1,
                                     autopct="%3.1f%%",
                                     radius = 1,
                                     pctdistance = 0.85,
                                     colors = outer_colors,
                                     textprops = dict(color="w"),
                                     wedgeprops = dict(width=0.3,edgecolor="w")) 

wedges2, texts2, autotexts2 = plt.pie(weight2,
                                     autopct="%3.1f%%",
                                     radius = 0.7,
                                     pctdistance = 0.75,
                                     colors = inner_colors,
                                     textprops = dict(color="w"),
                                     wedgeprops = dict(width=0.3,edgecolor="w")) 

plt.legend(wedges1,
          elements,
          fontsize = 12,
          title = "配料表",
          loc = "center left",
          bbox_to_anchor = (0.91,0,0.3,1))
# 通过legend函数中的loc和bbox_to_anchor参数控制图例位置

plt.setp(autotexts1,size = 15, weight = "bold")
plt.setp(autotexts2,size = 15, weight = "bold")

plt.setp(texts1,size = 12)
plt.title("不同果酱面包的配料比列例")

plt.show()

```
