```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,1000)
y1 = np.sin(x)

plt.plot(x,y1,ls="-",lw=1,color="blue",label="figure")

plt.legend(loc="upper left")

# 函数plt.text()添加图形内容的无指向性注释文本
plt.text(3,np.sin(3),"y=sin(x)",fontsize=18,weight="bold",color="r")  # 3表示横坐标，np.sin(3)表示注释的位置，后面的是注释的文本

# 添加图形的标签
plt.title("y=sin(x)",fontsize=18)
plt.xlabel("x",fontsize=18)
plt.ylabel("y",fontsize=18)

# 设置横坐标刻度
plt.xticks(np.arange(0,11,1))
plt.show()
```
